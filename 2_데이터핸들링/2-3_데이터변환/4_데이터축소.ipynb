```python
import pandas as pd 
from sklearn.preprocessing import StandardScaler 
from sklearn.decomposition import PCA 

df = pd.read_csv("../data/예제/USArrests_rownames.csv", index_col=0)
df.head() 
```

```python
x = StandardScaler().fit_transform(df)
```

```python
x
```

```python
pca = PCA(n_components=4)
pca_arr = pca.fit_transform(x)
```

```python
result = {"분산 기여율": pca.explained_variance_ratio_, 
          "누적 기여율": pca.explained_variance_ratio_.cumsum()}
```

```python
result = pd.DataFrame(result, index=["PCA1", "PCA2", "PCA3", "PCA4"])
```

```python
import pandas as pd 
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = pd.read_csv("../data/예제/USArrests_rownames.csv", index_col=0)
display(df.head())
```

```python
x = StandardScaler().fit_transform(df)

pca = PCA(n_components=4)

pca_arr = pca.fit_transform(x)

result = {
  "분산 기여율": pca.explained_variance_ratio_, 
  "누적 기여율": pca.explained_variance_ratio_.cumsum()
}

print(result)

result = pd.DataFrame(result, index=["PCA1", "PCA2", "PCA3", "PCA4"])
display(result)
```

```python
pca = PCA(n_components=2)
pca_arr = pca.fit_transform(x)
df_pca = pd.DataFrame(pca_arr, columns=["PCA1", "PCA2"])
df_pca
```
