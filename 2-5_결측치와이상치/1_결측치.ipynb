```python
# 이상치
obj = {
  'student_id': ['s1', 's2', 's3', 's4', 's5', 's6'], 
  'stat_score': [55, 90, 85, 10, 88, 99], 
  'math_score': [65, 99, 67, 70, 57, 80]
}

df = pd.DataFrame(obj) 

df
```

```python
q1 = df['stat_score'].quantile(0.25)
q3 = df['stat_score'].quantile(0.75)

iqr = q3 - q1 

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr 

print(lower)
print(upper)
```

```python
df[(df["stat_score"] < lower) | (df["stat_score"] > upper)]
```

```python
df[(df["stat_score"] > lower) & (df["stat_score"] < upper)]
```
